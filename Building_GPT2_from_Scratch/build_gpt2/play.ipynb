```python
from transformers import GPT2LMHeadModel
```

```python
model_hf = GPT2LMHeadModel.from_pretrained("/root/autodl-tmp/model")  # 实际加载的是124M版本的GPT-2模型，如果想要1558M版本的模型，可以用from_pretrained('gpt2-medium')
sd_hf = model_hf.state_dict()  # 状态字典也就是获取模型原始张量，即模型权重

for k, v in sd_hf.items():  # 由于就是一个字典，所以可以直接获取键值对，输出为GPT-2模型内的不同参数及形状
    print(k, v.shape)
```
